```python
import pandas as pd

# Load the CSV file
file_path = r"/content/cleaned.csv"
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
print(data.head())

# Check the data types and missing values
print(data.info())

# Summary statistics
print(data.describe())

```

```python
# Group by weather conditions and count the number of accidents
accidents_by_weather = data['Weather_conditions'].value_counts()

# Visualize the distribution of accidents by weather conditions
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
accidents_by_weather.plot(kind='bar')
plt.title('Distribution of Accidents by Weather Conditions')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
import folium

# Create a map centered around a specific location (for example, New York City)
m = folium.Map(location=[40.7128, -74.0060], zoom_start=10)

# Add markers for each accident location
for index, row in data.iterrows():
    folium.Marker([row['Accident_severity'], row['Accident_severity']], popup=row['Type_of_collision']).add_to(m)

# Display the map
m.save('accident_map.html')
```
